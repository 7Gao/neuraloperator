```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('fourier_m2_ic_e-2.npz')
pred = data['y_pred']
true = data['y_true']

l2_error = np.zeros(200)
for i in range(200):
    l2_error[i] = np.linalg.norm(pred[i, 0, :] - true[i, 0, :])
    l2_error[i] /= np.linalg.norm(true[i, 0, :])
print(np.mean(l2_error))
```

```python
plt.plot(pred[0, 1, :], label='Predicted')
plt.plot(true[0, 1, :], label='True')
plt.show()
```
