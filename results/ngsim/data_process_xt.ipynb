```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_rho = np.load('FNO_pred_NGSIM_xt_split30_rho_.npz')
data_rho_u = np.load('FNO_pred_NGSIM_xt_split30_rho_up.npz')
data_rho_d = np.load('FNO_pred_NGSIM_xt_split30_rho_down.npz')
data_rho_ud = np.load('FNO_pred_NGSIM_xt_split30_rho_updown.npz')

pred_rho = data_rho['y_pred']
pred_rho_u = data_rho_u['y_pred']
pred_rho_d = data_rho_d['y_pred']
pred_rho_ud = data_rho_ud['y_pred']
true_rho = data_rho['y_true']


error_rho = np.linalg.norm(pred_rho - true_rho, axis=(2, 3)) / np.linalg.norm(true_rho, axis=(2, 3))
error_rho_u = np.linalg.norm(pred_rho_u - true_rho, axis=(2, 3)) / np.linalg.norm(true_rho, axis=(2, 3))
error_rho_d = np.linalg.norm(pred_rho_d - true_rho, axis=(2, 3)) / np.linalg.norm(true_rho, axis=(2, 3))
error_rho_ud = np.linalg.norm(pred_rho_ud - true_rho, axis=(2, 3)) / np.linalg.norm(true_rho, axis=(2, 3))

print('rho error:', np.mean(error_rho))
print('rho error up:', np.mean(error_rho_u))
print('rho error down:', np.mean(error_rho_d))
print('rho error updown:', np.mean(error_rho_ud))
```

```python
# data = np.load('FNO_pred_NGSIM_xt.npz')
data = np.load('FNO_pred_NGSIM_xt_split30_rhou_updown.npz')
pred = data['y_pred']
true = data['y_true']

error = np.linalg.norm(pred - true, axis=(2, 3)) / np.linalg.norm(true, axis=(2, 3))
print('error:', np.mean(error))

index = 0
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

im0 = axs[0, 0].imshow(true[index, 0, :, :].T, vmin=0, vmax=.8, cmap='jet')
axs[0, 0].set_title('$\\rho$ true')
axs[0, 0].set_ylabel('x')
axs[0, 0].set_xlabel('t')
fig.colorbar(im0, ax=axs[0, 0])

im0 = axs[1, 0].imshow(pred[index, 0, :, :].T, vmin=0, vmax=.8, cmap='jet')
axs[1, 0].set_title('$\\rho$ pred')
axs[1, 0].set_ylabel('x')
axs[1, 0].set_xlabel('t')
fig.colorbar(im0, ax=axs[1, 0])

im1 = axs[0, 1].imshow(true[index, 1, :, :].T, vmin=0, vmax=.8, cmap='jet')
axs[0, 1].set_title('$u$ true')
axs[0, 1].set_ylabel('x')
axs[0, 1].set_xlabel('t')
fig.colorbar(im1, ax=axs[0, 1])

im1 = axs[1, 1].imshow(pred[index, 1, :, :].T, vmin=0, vmax=.8, cmap='jet')
axs[1, 1].set_title('$u$ pred')
axs[1, 1].set_ylabel('x')
axs[1, 1].set_xlabel('t')
fig.colorbar(im1, ax=axs[1, 1])

fig.suptitle('TFNO')

plt.show()
index += 1
```
