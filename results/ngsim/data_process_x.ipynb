```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('FNO_pred_NGSIM_rhou_ud_skip1_gap42.npz')
pred = data['y_pred']
true = data['y_true']

l2_error = np.zeros(229)
for i in range(229):
    l2_error[i] = np.linalg.norm(pred[i, :, :] - true[i, :, :])
    l2_error[i] /= np.linalg.norm(true[i, :, :])
print(np.mean(l2_error))
```

```python
index = 0
```

```python
plt.plot(pred[index, 1, :], label='Predicted')
plt.plot(true[index, 1, :], label='True')
plt.legend()
# plt.ylim(0, 0.5)
plt.show()
index += 1
```
